```python
# import library
import pandas as pd
import numpy as np
```

```python
# import csv file (patients)
fileNameStr = './patients.csv' 
Patients = pd.read_csv(fileNameStr,encoding = "ISO-8859-1",dtype = str) 
Patients
```

```python
# check data type
Patients.dtypes
```

```python
# check how many column, row in this dataset
Patients.shape
```

```python
Patients.isna()
```

```python
# check missing value
Patients.isnull().sum().sort_values(ascending=False)
```

```python
# drop column (SUFFIX, DEATHDATE, MAIDEN, PREFIX)
Patients = Patients.drop(columns = ['SUFFIX', 'DEATHDATE', 'MAIDEN','PREFIX'])
```

```python
Patients.shape
```

```python
# convert string to float
Patients['LAT'] = Patients['LAT'].astype('float')
Patients['LON'] = Patients['LON'].astype('float')
Patients['HEALTHCARE_EXPENSES'] = Patients['HEALTHCARE_EXPENSES'].astype('float')
Patients['HEALTHCARE_COVERAGE'] = Patients['HEALTHCARE_COVERAGE'].astype('float')
```

```python
# convert string to datetime
Patients['BIRTHDATE'] = pd.to_datetime(Patients['BIRTHDATE'])
```

```python
Patients.dtypes
```

```python
Patients.head()
```

```python
# Selecting duplicate rows based on list of column names, such as PII data 'SSN' 'ID'
duplicate_patients = Patients[Patients.duplicated(['Id', 'SSN'])]
 
print("Duplicate Rows based on ID and SSN :")
 
# Print the resultant Dataframe
duplicate_patients
```

```python
 # check All patients are living in Massachusetts state
Patients['STATE'].unique()
```

```python
# remove the digit in the first name and last name
Patients['FIRST'] = Patients['FIRST'].str.replace('\d+', '')
Patients['LAST'] = Patients['LAST'].str.replace('\d+', '')
```

```python
# add column full name (first and last name)
Patients['full_name']=Patients['FIRST'] +' ' + Patients['LAST']
Patients
```

```python
# groupby different column
grouped = Patients.groupby(by = ['GENDER', 'ETHNICITY', 'RACE'])
```

```python
grouped.apply(np.mean)[['HEALTHCARE_COVERAGE','HEALTHCARE_EXPENSES']]
```
